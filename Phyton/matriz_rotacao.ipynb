```python
import numpy as np
# Função que retorna a matriz de rotação em z
def Rz(theta):
  return np.array([
      [np.cos(theta), -np.sin(theta), 0],
      [np.sin(theta), np.cos(theta), 0],
      [0,          0,          1]
  ])
```

```python
theta = np.deg2rad(30)
Rz(theta)
```

```python
pb = np.array([3, 7, 0])
pa_org = np.array([10, 5, 0])
theta = np.deg2rad(30)
pa = Rz(theta) @ pb + pa_org
print(pa)
```
